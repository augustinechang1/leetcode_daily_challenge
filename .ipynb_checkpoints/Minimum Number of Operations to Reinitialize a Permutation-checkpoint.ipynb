```python
n = 4
```

```python
class Solution:
    def reinitializePermutation(self, n: int):
        # build out a list
        # iterate through list 
        # condition, change
        # count +1
        # repeat 
        
        a = list(range(n))
        
        b = [None] * n
        
        cnt = 0
        
        while a != b:
            
            for i in range(n):

                if i % 2 == 0:
                    
                    b[int(i)] = a[int(i/2)]
                    
                elif i % 2 == 1:
                    
                    b[int(i)] = a[int(n / 2 + (i - 1) / 2)]
                
                else:
                    break
            print(b)
            cnt += 1
            
        return cnt
```

```python
a = Solution()
```

```python
a.reinitializePermutation(n)
```
